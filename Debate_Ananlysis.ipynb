## **Extract *Country, Session, Year, Speech* from USA Sessions**

**Import header files**

```python
import glob
import pandas as pd
import re
```

**Convert all the folders into a list**

```python
folders = glob.glob('Converted sessions/Session*')
folders
```

**Define a dataframe to store *Country, Session, Year, Speech***

```python
df = pd.DataFrame(columns=('Country','Session','Year','Speech'))
```

**Extract the attributes by splitting the *filename*** 

```python
i = 0
for file in folders:
  speech = glob.glob(file+'/USA*.txt')
  with open(speech[0],encoding='utf-8') as f:
    df.loc[i,'Speech'] = f.read()
    df.loc[i,'Country'] = speech[0].split('_')[0].split('/')[-1]
    df.loc[i,'Session'] = speech[0].split('_')[1]
    df.loc[i,'Year'] = speech[0].split('_')[2].split('.')[0]
    i += 1
```

## **Text Cleaning (i.e. removing special and unecessary characters)**

**Defining function for cleaning**

```python
def clean(text):
  text = re.sub("[0-9]+.\t","",str(text))
  text = re.sub("\n"," ",str(text))
  text = re.sub("\n ","",str(text))
  text = re.sub("Mr\.","Mr",str(text))
  text = re.sub("Mrs\.","Mrs",str(text))
  text = re.sub("\"","",str(text))
  text = re.sub("'s","",str(text))
  text = re.sub("-"," ",str(text))
  text = re.sub("_ ","",str(text))
  text = re.sub("[\(\[].*?[\)\]]","",str(text))
  return text
```

**Cleaning the text**

```python
df['Clean Speech'] = df['Speech'].apply(clean)
df
```

## **Splitting Sentences**

**Defining function for splitting sentences on *period* or *question mark*** 

```python
def sentences(text):
  text = re.split('[.?]',text)
  clean_sent = []
  for sent in text:
    clean_sent.append(sent)
  return clean_sent
```

**Adding the sentence list to the dataframe**

```python
df['Sentence'] = df['Clean Speech'].apply(sentences)
df
```

```python
df['Sentence'][1]
```

**Creating a dataframe having *Year, Sentence, Length* for each sentence**

```python
row_list = []

for i in range(len(df)):
  for sent in df.loc[i,'Sentence']:
    row_list.append({'Year':df.loc[i,'Year'], 'Sentence':sent, 'Len':len(sent)})

df2 = pd.DataFrame(row_list)
```

## **Extracting sentences with president's mention**

**Importing *spacy* and *matcher* for text processing and pattern finding**

```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load('en_core_web_sm')
```

**Defining a pattern in our function**

```python
def find_names(text):

  matcher = Matcher(nlp.vocab)

  pattern = [{"LOWER":"president"},
             {"POS":"ADP","OP":"?"},
             {"POS":"PROPN"}]

  matcher.add("names",[pattern])
  matches = matcher(nlp(text))

  names = []

  for i in range(len(matches)):
    names.append(nlp(text)[matches[i][1]:matches[i][2]])

  return names
```

**Extracting president's mention in each sentence**

```python
df2["President_Names"] = df2["Sentence"].apply(find_names)
df2.head()
```

**Displaying the mentions**

```python
mentions = []

for i in range(df2.shape[0]):
  if df2["President_Names"][i] != []:
    mentions.append(df2["President_Names"][i])

mentions
```

## **Extracting sentences talking about schemes and campaigns**

**Function to check for schemes**

```python
def prog_sent(text):

  patterns = [r'\b(?i)'+'plan'+r'\b',
              r'\b(?i)'+'programme'+r'\b',
              r'\b(?i)'+'scheme'+r'\b',
              r'\b(?i)'+'campaign'+r'\b',
              r'\b(?i)'+'initiative'+r'\b',
              r'\b(?i)'+'conference'+r'\b',
              r'\b(?i)'+'agreement'+r'\b',
              r'\b(?i)'+'alliance'+r'\b']
  
  flag = 0

  for pat in patterns:
    if re.search(pat,text) != None:
      flag = 1
      break

  return flag

df2["Check_Schemes"] = df2["Sentence"].apply(prog_sent)

df2
```

**Function to extract schemes**

```python
def all_schemes(text,check):

  schemes = []

  prog_list = ['programme','scheme','initiative','campaign','agreement','conference','alliance','plan']

  pattern = [{'POS':'DET'},
             {'POS':'PROPN','DEP':'compound'},
             {'POS':'PROPN','DEP':'compound'},
             {'POS':'PROPN','OP':'?'},
             {'POS':'PROPN','OP':'?'},
             {'POS':'PROPN','OP':'?'},
             {'LOWER':{'IN':prog_list},'OP':'?'}]

  if check == 0:
    return schemes
  
  matcher = Matcher(nlp.vocab)
  matcher.add('scheme',[pattern])
  matches = matcher(nlp(text))

  for i in range(len(matches)):

    start,end = matches[i][1],matches[i][2]

    if nlp(text)[start].pos_ == 'DET':
      start += 1

    span = str(nlp(text)[start:end])

    if (len(schemes)!=0) and (schemes[-1] in span):
      schemes[-1] = span
    else:
      schemes.append(span)

  return schemes

df2["Schemes"] = df2.apply(lambda x:all_schemes(x.Sentence,x.Check_Schemes),axis=1)
```
